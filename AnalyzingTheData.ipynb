```python
# Imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset
from sklearn.preprocessing import LabelEncoder
```

```python
# This is the raw data from Kaggle (https://www.kaggle.com/datasets/uom190346a/sleep-health-and-lifestyle-dataset?resource=download)
originalData = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
print(originalData)
```

```python
originalData.columns
```

```python
# Begin Pre-Processing the Data
```

```python
# Dropping an unnecessary column
processedData = originalData.drop(['Person ID'], axis = 1)
processedData
```

```python
# 1-Hot Encoding the 'Sleep Disorder' Column (OPTIONAL)
#processedData = pd.concat([processedData, pd.get_dummies(processedData['Sleep Disorder'])], axis = 'columns')
#processedData = processedData.drop(['Sleep Disorder'], axis=1)
#processedData = processedData.rename(columns={'None': 'No Disorder'})
#processedData
```

```python
# Splitting 'Blood Pressure' column into 'Systolic Pressure' and 'Diastolic Pressure'
processedData[['Systolic Pressure', 'Diastolic Pressure']] = processedData['Blood Pressure'].str.split('/', expand=True)
processedData = processedData.drop(['Blood Pressure'], axis = 1)
processedData['Systolic Pressure'] = processedData['Systolic Pressure'].astype(int)
processedData['Diastolic Pressure'] = processedData['Diastolic Pressure'].astype(int)
processedData
```

```python
# Replace all 'Normal Weight' with 'Normal' in the 'BMI Category' column because they mean the same thing
processedData['BMI Category'] = processedData['BMI Category'].replace('Normal Weight', 'Normal')
processedData
```

```python
# Numerically Encoding the 'BMI Category' values (OPTIONAL)
#encoding = {'Normal': 1, 'Overweight': 2, 'Obese': 3}
#processedData['BMI Category'] = processedData['BMI Category'].map(encoding)
#processedData
```

```python
# Pre-Processing complete! Begin visualising the relationships between the features and target
```

```python
# Loading in all target/feature names to make comparison faster/easier
target = 'Quality of Sleep'
features = []

for col in processedData.columns:
    if col != 'Quality of Sleep':
        features.append(col)

print("Features: ", features)
```

```python
# Getting the correlation matrix
processedData.corr(numeric_only = 'True')
```

```python
# Create coloured bar chart to visualise the (avg) 'Quality of Sleep' of different professions
avgQOS = processedData.groupby('Occupation')['Quality of Sleep'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Occupation', y='Quality of Sleep', data=avgQOS, palette='husl')

plt.xticks(rotation=45)
plt.title('Average Quality of Sleep by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Average Quality of Sleep')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Create coloured scatterplot to visualise 'Quality of Sleep' by 'Gender'
plt.figure(figsize=(12, 2))

male = processedData[processedData['Gender'] == 'Male']
female = processedData[processedData['Gender'] == 'Female']

plt.plot(male['Quality of Sleep'], 'o', color='blue', label='Male')
plt.plot(female['Quality of Sleep'], 'o', color='pink', label='Female')

plt.gca().get_yaxis().set_visible(False)
plt.title('Quality of Sleep by Gender')
plt.xlabel('Quality of Sleep')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, axis='x')
plt.tight_layout()
plt.show()
```

```python
# Create coloured scatter-plot to visualise 'Quality of Sleep' by 'BMI Category'
plt.figure(figsize=(12, 2))

normalData = processedData[processedData['BMI Category'] == 'Normal']
overweightData = processedData[processedData['BMI Category'] == 'Overweight']
obeseData = processedData[processedData['BMI Category'] == 'Obese']

plt.plot(normalData['Quality of Sleep'], 'o', color='green', label='Normal')
plt.plot(overweightData['Quality of Sleep'], 'o', color='yellow', label='Overweight')
plt.plot(obeseData['Quality of Sleep'], 'o', color='red', label='Obese')

plt.gca().get_yaxis().set_visible(False)
plt.title('Quality of Sleep by BMI Category')
plt.xlabel('Quality of Sleep')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, axis='x')
plt.tight_layout()
plt.show()
```

```python
# Create coloured scatter-plot to visualise Quality of Sleep by 'Sleep Disorder'
plt.figure(figsize=(12, 2))

noneData = processedData[processedData['Sleep Disorder'] == 'None']
sleepApneaData = processedData[processedData['Sleep Disorder'] == 'Sleep Apnea']
insomniaData = processedData[processedData['Sleep Disorder'] == 'Insomnia']

plt.plot(noneData['Quality of Sleep'], 'o', color='black', label='None')
plt.plot(sleepApneaData['Quality of Sleep'], 'o', color='blue', label='Sleep Apnea')
plt.plot(insomniaData['Quality of Sleep'], 'o', color='red', label='Insomnia')

plt.gca().get_yaxis().set_visible(False)
plt.title('Quality of Sleep by Sleep Disorder')
plt.xlabel('Quality of Sleep')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, axis='x')
plt.tight_layout()
plt.show()
```

```python
# Generating scatter-plots to visualise the rest of the features vs Quality of Sleep
for feat in features:
    if feat != 'Occupation' and feat != 'Gender' and feat != 'BMI Category' and feat != 'Sleep Disorder':
        plt.figure(figsize=(12, 6))
        plt.plot(processedData[feat], processedData['Quality of Sleep'], 'o')
        plt.xlabel(feat)
        plt.ylabel('Quality of Sleep')
        plt.title(feat + ' vs Quality of Sleep')
        plt.show()
```
